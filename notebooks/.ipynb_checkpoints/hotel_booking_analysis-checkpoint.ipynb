```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../data/hotel_bookings.csv")
```

```python
df.info()
```

```python
df['children'] = df['children'].fillna(0).astype(int)
```

```python
df.isnull().sum()
```

### Data Cleaning

```python
df['children']=df['children'].fillna(0)
df['adults']=df['adults'].fillna(0)
df['babies']=df['babies'].fillna(0)
```

```python
df['country']=df['country'].fillna('unknown')
```

```python
df['agent']=df['agent'].fillna(0)
```

```python
df.drop(columns=['company'],inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
# hotel booking
df['hotel'].value_counts()
```

```python
# Hotel Cancelation
df['is_canceled'].value_counts(normalize='True') * 100
```

```python
# monthly booking trend
df['arrival_date_month'].value_counts()
```

```python
# average daily rate
df['adr'].mean()
```

```python
# top 10 countries by bookings
df['country'].value_counts().head(10)
```

```python
#hotel tyoe vs cancellation
df.groupby('hotel') ['is_canceled'].mean() *100
```

```python
#Lead time vs canceletation
df.groupby('is_canceled')['lead_time']. mean() 
```

```python
#Create new columns
df['total_guests']=df['adults']+df['children']+df['babies']
df['total_nights']=df['stays_in_week_nights']+df['stays_in_weekend_nights']

total_guest_number=df['total_guests'].sum()
print("Total Guests:",total_guest_number)

total_nights_number=df['total_nights'].sum()
print("Total Nights:",total_nights_number)
```

## Visualization

```python
# Hotel by distribution
sns.countplot(x='hotel', data=df)
plt.title("Hotel Type")
plt.show()

#Cancelation status
sns.countplot(x='is_canceled', data=df)
plt.title("Booking vs Cancellation")
plt.show()
```

```python
#Monthly Bookings
months_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]
df['arrival_date_month'].value_counts().reindex(months_order).plot(kind='bar')
plt.title("Monthly Bookings")
plt.show()
```

```python
#Booking By Market Segment
sns.countplot(x='market_segment', data=df)
plt.xticks(rotation=45)
plt.title("Booking by Market Segment")
plt.show()
```

```python
# Lead Time Contribution 
sns.histplot(df['lead_time'], bins=50)
plt.title("Lead Time Contribution")
plt.show()

# Top Countries
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 countries")
plt.show()
```

```python
# Average Daily Rate (ADR) by hotel type
sns.boxplot(x='hotel', y='adr',data=df)
plt.show()
```
